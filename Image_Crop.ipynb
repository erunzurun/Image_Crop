```python
def DeleteAllFiles(filePath):
    if os.path.exists(filePath):
        for file in os.scandir(filePath):
            os.remove(file.path)
        return 'Remove All Files'
    else:
        return 'Directory Not Found'
```

# xml_first

```python
img_dir = ""
xml_dir = ""
save_dir = ""
save_folder_name = ""

img_list = os.listdir(img_dir)
xml_list = os.listdir(xml_dir)
name_list = os.listdir(save_folder_name)
```

```python
name_list
```

```python
print(len(img_list))
```

```python
#경로 이미지 전부 삭제
# for c in name_list:
#     DeleteAllFiles(f"{save_folder_name}/{c}")

for i in range(len(img_list)):
    full_path = img_dir + img_list[i]
    img_array = np.fromfile(full_path, np.uint8)
    img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
    
    h, w, c = img.shape
    
    save_name = img_list[i].replace('.jpg', '')
    
    if img is None:
        print('Image load failed')
        sys.exit()
    
    with open(xml_dir + xml_list[i], "r", encoding='utf-8') as xml:
        tree = Et.parse(xml)
        root = tree.getroot()

    for x in root:
        count = 1
        name = ""
        x1 = ""
        y1 = ""
        x2 = ""
        y2 = ""
    #     print(x.tag, x.attrib, x.text)
        for y in x:
            count = 1
            attrib_y = y.attrib
    #         print(qwe)
            while attrib_y:
                name = attrib_y['name']
                break
    #         print(y.tag, y.attrib, y.text)
            for z in y:
                attrib_z = z.attrib
                while attrib_z:
                    x1 = float(attrib_z['x1']) * w 
                    x2 = float(attrib_z['x2']) * w
                    y1 = float(attrib_z['y1']) * h
                    y2 = float(attrib_z['y2']) * h
                    break
                x1 = int(round(x1, 0))
                x2 = int(round(x2, 0))
                y1 = int(round(y1, 0))
                y2 = int(round(y2, 0))
                         
                print(x1, x2, y1, y2)
                crop_img = img[y1:y2, x1:x2]
                
                crop_img_name = img_list[i]
                extension = os.path.splitext(crop_img_name)[1] # 이미지 확장자\

#                 print(crop_img_name)
#                 print(extension)

                result, encoded_img = cv2.imencode(extension, crop_img)
    
                if result:
                    while os.path.exists(f"{save_dir}/{name}/{save_name}_{name}_{count}.jpg"):
                        count += 1
                    with open(f"{save_dir}/{name}/{save_name}_{name}_{count}.jpg", mode='w+b') as f:
                        encoded_img.tofile(f)

                if result:
                    if not os.path.exists(f"{save_dir}/{name}/{save_name}_{name}_{count}.jpg"):
                        with open(f"{save_dir}/{name}/{save_name}_{name}_{count}.jpg", mode='w+b') as f:
                            encoded_img.tofile(f)
                    else:
                        count += 1
                        with open(f"{save_dir}/{name}/{save_name}_{name}_{count}.jpg", mode='w+b') as f:
                            encoded_img.tofile(f)
                            
#                 cv2.imshow("cropped", crop_img)
#                 cv2.waitKey(0)

```
